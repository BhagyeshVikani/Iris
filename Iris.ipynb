```python
# importing necessary library
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading Iris dataset and setting its index to "Id"
original_data = pd.read_csv("Iris.csv")
original_data = original_data.set_index("Id", drop = True)
original_data.head()
```

```python
# column names in Iris dataset except "Species"
# only feature columns
column_names = list(original_data.columns.values)
column_names.remove("Species")
column_names
```

```python
# number of counts for all label classes
original_data["Species"].value_counts()
```

```python
# mapping "Species" column with number
modified_data = original_data
modified_data['Species'] = modified_data['Species'].map({'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2})
modified_data.head()
```

```python
# plot for feature vs "Species"
modified_data[["SepalLengthCm", "Species"]].plot(kind = "scatter", x = "Species", y = "SepalLengthCm")
modified_data[["SepalWidthCm", "Species"]].plot(kind = "scatter", x = "Species", y = "SepalWidthCm")
modified_data[["PetalLengthCm", "Species"]].plot(kind = "scatter", x = "Species", y = "PetalLengthCm")
modified_data[["PetalWidthCm", "Species"]].plot(kind = "scatter", x = "Species", y = "PetalWidthCm")
```

```python
# grouping dataset by "Species"
grouped_data = modified_data.groupby(by = "Species")
grouped_data[column_names].mean()
```

```python
# plot for mean of each feature for each label class
grouped_data[column_names].mean().plot(kind = "barh", stacked = True)
```
